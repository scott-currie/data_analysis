## Video Game Sales

```python
import pandas as pd
import numpy as np
import datetime as dt

```

```python
df = pd.read_csv('./vgsales.csv')
```

```python
df.head(10)
```

### 1. Which company is the most common video game publisher?- 
- Get a series of all publishers
- Group by publisher name
- Get the max

```python
publishers = df.Publisher.groupby(df.Publisher)
```

```python
publishers.head()
```

```python
publishers.count().sort_values(ascending=False)
```

### Electronic Arts appears 1351 times.

### 2. What is the most common platform?
- Get series of platforms
- Group by platform
- Sort descending

```python
platforms = df.Platform.groupby(df.Platform)
```

```python
platforms.count().sort_values(ascending=False)
```

### Nintendo DS is the most common platform with 2163 titles.

### 3. What is the most common genre?
- Get a series of genres
- Group by genre
- Sort descending

```python
genres = df.Genre.groupby(df.Genre)
```

```python
most_common_genre = df['Genre'].value_counts().idxmax()
```

```python
print(f'Most common genre: { most_common_genre }')
```

### Action genre is most common with 3316 titles.

### 4. What are the top 20 highest grossing games?

### 5. For North American video game sales, what’s the median?

### 6. For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_na_seller = df.sort_values(by=['NA_Sales'], ascending=False).NA_Sales[0]
```

```python
top_na_seller
```

```python
na_sales_mean = df.NA_Sales.mean()
```

```python
na_sales_mean
```

```python
na_sales_std = df.NA_Sales.std()
```

```python
na_sales_std
```

```python
print(f'Wii Sports is { (top_na_seller - na_sales_mean) / na_sales_std } standard deviations above the mean.')
```
